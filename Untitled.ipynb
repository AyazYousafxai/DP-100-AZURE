```python
import pandas as pd
```

```python
from sklearn.linear_model import LogisticRegression
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('diabetes.csv')
```

```python
feature = data.columns
```

```python
len(feature)
```

```python
X,feature =feature[-1],feature[1:-1]
```

```python
feature,X
```

```python
for i in feature:
    data.boxplot(column=i,by=X,figsize=(6,6))
    plt.title(i)
plt.show()
```

```python
x,y=data[feature].values,data[X].values
```

```python
data[feature]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=123,test_size=0.30)
```

```python
ytest
```

```python
req=0.01
model=LogisticRegression(C=1/req,solver='liblinear')
```

```python
pred = model.fit(xtrain,ytrain)
```

```python
p=pred.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import precision_score,recall_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import roc_curve,auc
```

```python
num = [i for i in range(0,7)]
cat=[7]
```

```python
cat
```

```python
len(num)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.pipeline import Pipeline
```

```python
scaler = Pipeline(steps=[
    ('scaler',StandardScaler())
])
encoder = Pipeline(steps=[
    ('encoder',OneHotEncoder(handle_unknown='ignore'))
])

pre_pipeline=ColumnTransformer(
    transformers=[
        ('num',scaler,num),
        ('cat',encoder,cat)
    ]
)
model1 = Pipeline(steps=[('preprocess',pre_pipeline),
                       ('model_logistic',LogisticRegression(C=1/req,solver='liblinear'))])
```

```python
print(model1)
```

```python
len(xtrain[0])
```
