```python
import pandas as pd
```

```python
!wget https://raw.githubusercontent.com/MicrosoftDocs/mslearn-introduction-to-machine-learning/main/Data/ml-basics/grades.csv
```

```python
data = pd.read_csv('grades.csv')
```

```python
data=data.dropna(axis=0,how='any')
```

```python
data.head()
```

```python
pass_student=pd.Series(data['Grade']>=60)
```

```python
data=pd.concat([data,pass_student.rename('Pass')],axis=1)
```

```python
from matplotlib import pyplot as plt
```

```python
fig=plt.figure(figsize=(10,8))
plt.bar(x=data.Name,height=data.Grade,color='orange')
plt.title('Student Grade')
plt.xlabel('Name')
plt.ylabel('Grade')
plt.grid(color='red',linestyle='--',linewidth=1,axis='y',alpha=0.6)
plt.xticks(rotation=120)
plt.show()
```

```python
fig, axis = plt.subplots(1,2, figsize=(10,4))
axis[0].bar(x=data.Name,height=data.Grade,color='orange')
axis[0].set_title('Student Grades')
axis[0].set_xticklabels(data.Name,rotation=120)

pass_count=data.Pass.value_counts()
axis[1].pie(pass_count,labels=pass_count)
axis[1].set_title('Pass count')
axis[1].legend(pass_count.keys().tolist())

fig.suptitle('student Data')
fig.show()
```

```python
grade= data['Grade']
fig = plt.figure(figsize=(8,3))
plt.hist(grade)
plt.title('student Grade')
plt.xlabel('Grade')
plt.ylabel('frequency')
fig.show()
```

```python
min_val = grade.min()
max_val = grade.max()
men_val = grade.mean()
mod_val = grade.mode()[0]
med_val = grade.median()
```

```python
fig = plt.figure(figsize=(10,4))
plt.hist(grade)
plt.axvline(x=min_val,color='orange',linestyle='dashed',linewidth=1)
plt.axvline(x=max_val,color='red',linestyle='dashed',linewidth=1)
plt.axvline(x=men_val,color='cyan',linestyle='dashed',linewidth=1)
plt.axvline(x=mod_val,color='yellow',linestyle='dashed',linewidth=1)
plt.axvline(x=med_val,color='gray',linestyle='dashed',linewidth=1)

plt.title('Data Distribution')
plt.xlabel('value')
plt.ylabel('frequ')
fig.show()
```

```python
fig = plt.figure(figsize=(10,4))
plt.boxplot(grade)
plt.title('Data Distribution')
fig.show()
```

```python
def show_dist(var_data):
    var = var_data
    fig, axis = plt.subplots(2,1,figsize=(10,3))
    
    axis[0].axvline(x=min_val,color='cyan',linestyle='dashed',linewidth=2)
    axis[0].axvline(x=max_val,color='red',linestyle='dashed',linewidth=2)    
    axis[0].axvline(x=mod_val,color='yellow',linestyle='dashed',linewidth=2)    
    axis[0].axvline(x=med_val,color='gray',linestyle='dashed',linewidth=2)  
    axis[0].axvline(x=men_val,color='orange',linestyle='dashed',linewidth=2)  
    
    axis[1].boxplot(var)
    axis[1].set_xlabel('value')
    
    fig.suptitle('Data Distribution')
    fig.show()
    
var = data['Grade']
show_dist(var)
```

```python
def show_density(var):
    fig=plt.figure(figsize=(10,4))
    
    var.plot.density()
    
    plt.title('density')
    
    plt.axvline(x=min_val,color='cyan',linestyle='dashed',linewidth=2)
    plt.axvline(x=max_val,color='red',linestyle='dashed',linewidth=2)    
    plt.axvline(x=mod_val,color='yellow',linestyle='dashed',linewidth=2)    
    plt.axvline(x=med_val,color='gray',linestyle='dashed',linewidth=2)  
    plt.axvline(x=men_val,color='orange',linestyle='dashed',linewidth=2)
    
    plt.show()
    
var = data['Grade']

show_density(var)
```
